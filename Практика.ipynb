Проверяем, что `0.95` нормального распределения лежит в (примерно) рамках $[\mu - \sigma, \mu + \sigma]$  

```python
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt

# Пример
loc, scale = 3, 10
n = 10000
X = norm(loc, scale).rvs(n)

# Рисуем график этого нормального распределения N(3, 10^2)
plt.hist(X, bins=100)

# Среднее 
plt.vlines(x=loc, ymin=0, ymax=350, linewidth=2, color='r');

# Отмечаем предполагаемые линии
plt.vlines(x=loc + 2*scale, ymin=0, ymax=350, linewidth=2, color='r');
plt.vlines(x=loc - 2*scale, ymin=0, ymax=350, linewidth=2, color='r');

# Ищем точки где куммулятивное норм распределение занимает 0.025 и 0.975 всей площади
plt.vlines(x=norm(loc, scale).ppf(0.025), ymin=0, ymax=350, linewidth=2, color='g');
plt.vlines(x=norm(loc, scale).ppf(0.975), ymin=0, ymax=350, linewidth=2, color='g');
```

Получается, что да, действительно это примерно так. Данный факт помогает решать задачи без необходимости считать обратную кумулятивную функцию плотности нормального распределения. Например для задач, где нужно использовать ЦПТ. Меня такое на собесе в Яндекс спрашивали.
